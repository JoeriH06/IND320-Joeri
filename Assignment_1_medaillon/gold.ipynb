```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load data from the Silver layer
silver_in = Path("../Data_assignment_1/silver/open_meteo_clean.csv")
data = pd.read_csv(silver_in)
```

```python
print("Dtypes:\n", data.dtypes, "\n")
print("Nulls per column:\n", data.isna().sum(), "\n")
display(data.head(3))
```

```python

# Convert time column to datetime
data["time"] = pd.to_datetime(data["time"], errors="coerce")

# Plot each numeric column against time
for column in data.columns:
    # skip the time column itself
    if column != "time":
        plt.figure(figsize=(10, 4))
        plt.plot(data["time"], data[column])
        plt.title(column)
        plt.xlabel("Time")
        plt.ylabel(column)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```
